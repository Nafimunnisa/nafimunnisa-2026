### Python Variable Naming Rules:

1.  **Must start with a letter or an underscore (_)**: A variable name cannot start with a number.
2.  **Can only contain alpha-numeric characters and underscores (a-z, A-Z, 0-9, and _)**: No spaces, hyphens, or other special characters are allowed.
3.  **Are case-sensitive**: `age`, `Age`, and `AGE` are three different variables.
4.  **Cannot be Python keywords**: Words like `if`, `else`, `for`, `while`, `print`, `class`, `def`, `return`, etc., are reserved and cannot be used as variable names.

Here are 5 examples of invalid variable names and why they are invalid:

1.  **`1variable`**
    *   **Reason**: Starts with a number. (Rule 1 violated)

2.  **`my-variable`**
    *   **Reason**: Contains a hyphen (`-`). Only underscores are allowed as special characters. (Rule 2 violated)

3.  **`my variable`**
    *   **Reason**: Contains a space. (Rule 2 violated)

4.  **`class`**
    *   **Reason**: `class` is a reserved Python keyword. (Rule 4 violated)

5.  **`@variable`**
    *   **Reason**: Starts with a special character (`@`) that is not an underscore. (Rule 1 and 2 violated)
