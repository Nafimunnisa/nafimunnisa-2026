```python
import math

# Principal amount
P = 1000
# Annual interest rate (5%)
R = 0.05
# Number of times interest is compounded per year (e.g., 4 for quarterly)
N = 4
# Number of years
T = 10

# Calculate future value using the compound interest formula
A = P * math.pow((1 + R/N), (N*T))

print(f"Principal Amount (P): ${P}")
print(f"Annual Interest Rate (R): {R*100}%")
print(f"Compounding Frequency per Year (N): {N}")
print(f"Number of Years (T): {T}")
print(f"Future Value (A) after {T} years: ${A:.2f}")
```

```python
import math

# Define the radius of the circle
radius = 7

# Calculate the circumference
circumference = 2 * math.pi * radius

print(f"The radius of the circle is: {radius}")
print(f"The circumference of the circle is: {circumference:.2f}")
```
