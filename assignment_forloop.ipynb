```python
try:
    num = int(input("Enter a number to see its multiplication table: "))
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
text = input("Enter a string to count its vowels: ")
vowel_count = 0
vowels = "aeiouAEIOU"

for char in text:
    if char in vowels:
        vowel_count += 1

print(f"The number of vowels in the string is: {vowel_count}")
```

```python
try:
    num = int(input("Enter a non-negative integer to calculate its factorial: "))

    if num < 0:
        print("Factorial does not exist for negative numbers.")
    elif num == 0:
        print("The factorial of 0 is 1.")
    else:
        factorial = 1
        for i in range(1, num + 1):
            factorial *= i
        print(f"The factorial of {num} is {factorial}.")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
names_input = input("Enter names separated by spaces (e.g., 'Alice Bob Eve'): ")
names_list = names_input.split()

e_count = 0
for name in names_list:
    for char in name:
        if char.lower() == 'e':
            e_count += 1

print(f"The letter 'e' appears {e_count} times in the provided names.")
```
